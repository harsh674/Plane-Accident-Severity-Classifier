## Now I am Going To use Random Forest Because random forest is that it is more accurate than the decision trees due to the reduction in the over-fitting.

```python
from sklearn.model_selection import train_test_split
```

```python
X=airplane_df[['Safety_Score', 'Days_Since_Inspection', 'Total_Safety_Complaints', 'Control_Metric','Turbulence_In_gforces','Cabin_Temperature','Accident_Type_Code','Max_Elevation','Violations','Adverse_Weather_Metric']]  # Features
y=airplane_df['Severity']
#train test split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn import metrics
```

```python
from sklearn.model_selection import cross_val_score
```

### We have found that in range of 100 to 250 of n_estimators, the model performs better,Because More Number of n_estimators will effect the performance of the model

```python
i=100
j=0
accuracy_arr = np.zeros(shape=(16,2))
while i<251:
    clf=RandomForestClassifier(n_estimators=i)

    #Train the model using the training sets y_pred=clf.predict(X_test)
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    accuracy_arr[j][0] = i
    accuracy_arr[j][1] = metrics.accuracy_score(y_test, y_pred)
    i=i+10
    j=j+1
print(accuracy_arr)    
```

### We found that at n_estimator = 200 or 210 model performs best

```python

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=200)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

```

```python
y_pred=clf.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

### Cross Validation Score of The Model

```python
cross_val_score(clf, X_train, y_train, cv=10)  
```
